```python
# Programa de uma lanchonete.
# Exibe um menu de produtos, permite que o cliente escolha os itens e as quantidades,
# calcula o valor total da compra e encerra quando a opção "SAIR" é selecionada.

print("LANCHONETE")
print('1 - Coxinha RS 5,00')
print('2 - Pastel RS 7,00')
print('3 - Café RS 4,00')
print('4 - Suco RS 6,00')
print('5 - SAIR')

total = 0
while True:
  op = int(input("Qual item gostaria de comprar?"))
  # The initial 'qtd' input here will be overwritten if op is 1-4.
  # It's better to ask for 'qtd' inside each specific item's 'if' or 'elif' block, or after selecting the item.
  # For now, I'm commenting it out as it's redundant.
  # qtd = int(input(" quantas unidades quer comprar?"))

  if (op == 1):
      qtd = int(input(" quantas unidades quer comprar?"))
      total = total + (qtd * 5.00)
  elif (op == 2):
      qtd = int(input(" quantas unidades quer comprar?"))
      total = total + (qtd * 7.00)
  elif (op == 3):
      qtd = int(input(" quantas unidades quer comprar?"))
      total = total + (qtd * 4.00)
  elif (op == 4):
      qtd = int(input(" quantas unidades quer comprar?"))
      total = total + (qtd * 6.00)
  elif (op == 5):
      print(f"Total da compra: R$ {total:.2f}") # Print total before breaking
      break
  else:
      print("Produto inválido. Selecionar outro.")

```
