```python
# Programa que calcula o valor da conta de energia elétrica.
# O usuário informa o consumo em kWh e o tipo de instalação
# (R - Residencial, C - Comercial ou I - Industrial).
# O programa aplica a tarifa correspondente e exibe o total a pagar.

kwh = float(input('Quantos kWh? '))
tipo = input('Qual o tipo da instalação? (R, C ou I) ')

if tipo == 'R':
    if kwh >= 500:
        preco = 0.65
    else:
        preco = 0.40

    print(f'Total a pagar: R$ {kwh * preco:.2f}')

elif tipo == 'C':
    if kwh >= 1000:
        preco = 0.60
    else:
        preco = 0.55

    print(f'Total a pagar: R$ {kwh * preco:.2f}')

elif tipo == 'I':
    if kwh >= 5000:
        preco = 0.60
    else:
        preco = 0.55

    print(f'Total a pagar: R$ {kwh * preco:.2f}')

else:
    print('Instalação inválida. Encerrando...')
```
